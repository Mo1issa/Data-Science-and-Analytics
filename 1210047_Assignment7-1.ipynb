## Defining the modules that will be used

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

### Data Preprocessing Steps

1. **Removing Irrelevant or Constant Features**
   - Dropped the `Id` column as it does not contribute to the model.


```python

# 1. Removing irrelevant or constant features
irrelevant_features = ['Id', 'Utilities', 'Street', 'PoolQC', 'Condition2', 'RoofMatl', 'MiscFeature']
df = df.drop(columns=irrelevant_features)

# Verify if there are any constant columns left
constant_columns = [col for col in df.columns if df[col].nunique() == 1]
df = df.drop(columns=constant_columns)
df
```


2. **Verifying Data Types**
   - Ensured that numerical features are correctly identified as numeric and ordinal features are recognized appropriately.


```python
# 2. Verifying data types
print(df.dtypes)
```

3. **Removing Duplicates**
   - Checked and removed any duplicate rows in the dataset.

```python
# 3. Removing duplicates
df = df.drop_duplicates()
df.duplicated().sum()
```

4. **Defining Feature Matrix and Target Vector**
   - Selected `OverallQual`, `YearBuilt`, `TotalBsmtSF`, `GrLivArea`, and `GarageCars` as features.
   - Defined `SalePrice` as the target variable.


```python
# 4. Defining feature matrix and target vector
features = ['OverallQual', 'YearBuilt', 'TotalBsmtSF', 'GrLivArea', 'GarageCars', 'MSZoning', 'Neighborhood', 'HouseStyle']
target = 'SalePrice'
X = df[features]
y = df[target]
X.info()
```

5. **Splitting the Data to Train and Test Sets**
   - Split the data into 80% training and 20% testing sets using a specified `random_state` (The ID)for reproducibility.


```python
# 5. Splitting the data to train and test sets
student_id = 1210047  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=student_id)
X_train
```

6. **Handling Missing Values**
   - Applied median imputation for missing values in numerical features.
   - Used the most frequent value imputation for missing values in ordinal features.


```python
# Define list of categorical features
cat_cols = X_train.select_dtypes(include=['object', 'category']).columns

# Define list of numerical features
num_cols = X_train.select_dtypes(include=['number']).columns

# Instantiate the imputer with the desired strategy
impute_na = SimpleImputer(strategy='constant', fill_value='NA')
impute_median = SimpleImputer(strategy='median')

# Fit the imputer object on the categorical training data
X_train[cat_cols] = impute_na.fit_transform(X_train[cat_cols])
X_test[cat_cols] = impute_na.transform(X_test[cat_cols])

# Fit the imputer object on the numerical training data
X_train[num_cols] = impute_median.fit_transform(X_train[num_cols])
X_test[num_cols] = impute_median.transform(X_test[num_cols])

```

7. **Encoding Categorical Data**
   - Encoded ordinal features using ordinal encoding.

```python
# Define the numerical, ordinal, and nominal features
numeric_features = ['YearBuilt', 'TotalBsmtSF', 'GrLivArea']
ordinal_features = ['OverallQual', 'GarageCars']
nominal_features = ['MSZoning', 'Neighborhood', 'HouseStyle']

# Separate features
X_numeric_train = X_train[numeric_features]
X_ordinal_train = X_train[ordinal_features]
X_nominal_train = X_train[nominal_features]

# Encode ordinal features
encoder_ord = OrdinalEncoder()
X_ordinal_train_encoded = encoder_ord.fit_transform(X_ordinal_train)

# Encode nominal features using one-hot encoding
encoder_nom = OneHotEncoder(handle_unknown='ignore', sparse_output=False) 
X_nominal_train_encoded = encoder_nom.fit_transform(X_nominal_train)
```

8. **Feature Scaling**
   - Scaled numerical features to standardize the range of values.


```python
import numpy as np
# Scale numerical features
scaler = StandardScaler()
X_numeric_train_scaled = scaler.fit_transform(X_numeric_train)

# Concatenate all preprocessed features
X_train_processed = np.concatenate([X_numeric_train_scaled, X_ordinal_train_encoded, X_nominal_train_encoded], axis=1)

# Repeat preprocessing for the test set
X_numeric_test = X_test[numeric_features]
X_ordinal_test = X_test[ordinal_features]
X_nominal_test = X_test[nominal_features]

X_numeric_test_scaled = scaler.transform(X_numeric_test)
X_ordinal_test_encoded = encoder_ord.transform(X_ordinal_test)
X_nominal_test_encoded = encoder_nom.transform(X_nominal_test)

X_test_processed = np.concatenate([X_numeric_test_scaled, X_ordinal_test_encoded, X_nominal_test_encoded], axis=1)
```

1. Fit a Linear Regression Model on the Transformed Data

```python
from sklearn.linear_model import LinearRegression

# Initialize the linear regression model
model = LinearRegression()

# Fit the model on the transformed training data
model.fit(X_train_processed, y_train)
```

2. Print the Model Parameters and Provide an Interpretation

```python
# Print the coefficients and intercept
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Mapping feature names to coefficients for interpretation
feature_names = (numeric_features + 
                 ordinal_features + 
                 encoder_nom.get_feature_names_out(nominal_features).tolist())

coefficients = model.coef_

# Print the feature names and their corresponding coefficients
for feature, coef in zip(feature_names, coefficients):
    print(f"{feature}: {coef}")
```

* Explanation:
### Interpretation:

* Intercept: The intercept is the expected mean value of the target (SalePrice) when all predictors are set to zero.
* Coefficients: Each coefficient represents the change in the target (SalePrice) for a one-unit change in the respective feature, holding all other features constant.

3. Evaluate the Model's Quality Using Appropriate Metrics

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predict on the training set
y_train_pred = model.predict(X_train_processed)

# Predict on the test set
y_test_pred = model.predict(X_test_processed)

# Calculate and print the metrics
train_mse = mean_squared_error(y_train, y_train_pred)
train_r2 = r2_score(y_train, y_train_pred)

test_mse = mean_squared_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)

print(f"Training MSE: {train_mse}")
print(f"Training R2: {train_r2}")

print(f"Test MSE: {test_mse}")
print(f"Test R2: {test_r2}")
```

4. Determine if the Model Exhibits Overfitting or Underfitting
Overfitting vs. Underfitting:

* Overfitting: The model performs well on the training data but poorly on the test data. This is indicated by a significantly lower MSE and higher R² score on the training set compared to the test set.
* Underfitting: The model performs poorly on both the training and test data, indicated by high MSE and low R² scores on both sets.
### Analysis:

* Compare the MSE and R² scores on the training and test sets.
* If the test MSE is much higher than the training MSE and the test R² is much lower than the training R², the model may be overfitting.
* If both MSEs are high and both R² scores are low, the model may be underfitting.
