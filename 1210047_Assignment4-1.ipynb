4. Select two numerical features and show their descriptive statistics, add your interpretation on each feature. Does they have skewness? Interpret.
- To show the descriptive statistics of two numerical features from the "Adult Income Dataset" and interpret their characteristics, I selected the desired numerical features. And choose the "age" and "hours_per_week" features for this analysis.



```python
# Select the numerical features
numerical_features = ["age", "capital-gain"]

# Show descriptive statistics for the selected features
numerical_stats = df[numerical_features].describe()

# Calculate skewness for each feature
skewness = df[numerical_features].skew()

# Print descriptive statistics
print("Descriptive Statistics for Numerical Features:")
print(numerical_stats)

# Print skewness for each feature
print("\nSkewness for each Numerical Feature:")
print(skewness)

```

* Age : The skewness value of approximately 0.56 indicates a slightly right-skewed distribution. This means that there are more individuals with ages higher than the median age compared to those lower than the median.
* capital-gain : The skewness value of approximately 11.89 indicates a highly right-skewed distribution. This means that there are very few individuals with high capital gains, leading to a long tail on the right side of the distribution.
* The skewness values confirm that both the "age" and "capital-gain" features exhibit right-skewed distributions


* I applaied the histogram for both age and capital-gain features to explain the skewness more and more so it will be clear :

```python
# Set up the figure and axes
figure, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the histogram for 'age'
sns.histplot(df['age'], ax=axes[0], kde=True, color='skyblue')
axes[0].set_title('Distribution of Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

# Plot the histogram for 'capital-gain'
sns.histplot(df['capital-gain'], ax=axes[1], kde=True, color='salmon')
axes[1].set_title('Distribution of Capital Gain')
axes[1].set_xlabel('Capital Gain')
axes[1].set_ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

* from the histogram above we can see clearly the skewness to the right in both age and Capital Gain 

5. Select two categorical features and verify if the data is balanced between the categories of each feature.
- By testing the calculation of frequency counts or proportions of each category in the dataset. If the counts or proportions are roughly equal across categories, we can consider the data to be balanced.
- I used the method Value_counts to check the balance between (education ,marital-status)

```python


# Select two categorical features
categorical_features = ["education", "marital-status"]

# Verify balance for each categorical feature
for feature in categorical_features:
    print("Balance for", feature)
    print(df[feature].value_counts(normalize=True))
    print()
```

* These findings suggest that the data is not perfectly balanced for both the "education" and "marital-status" features. Some categories are more prevalent than others, indicating an imbalance in the distribution of data across categories.Therefore,it is approximately balanced as there is no extreme skewness or dominance of one category over others.But in the  future analysis or modeling may be needed to address any potential biases introduced by this imbalance.














5. Based on your intuition, select two features that you think they could be correlated. Verify your assumption using both visualization and the correlation coefficient approaches.
- we need to steps to answer this Quastion :

* Step one : visualize the relationship between these two features 'educational-num' and 'income'
 using a scatter plot:



```python

# Scatter plot to visualize the relationship between 'age' and 'hours-per-week'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='hours-per-week', data=df)
plt.title('Relationship between Age and Hours Worked per Week')
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.show()
```

* Step two : calculate the correlation coefficient between these two features:



```python
# Calculate the correlation coefficient
correlation = df['age'].corr(df['hours-per-week'])
print("Correlation coefficient between 'age' and 'hours-per-week':", correlation)
```

* Based on the correlation coefficient of approximately 0.072 between "age" and "hours-per-week," it suggests a very weak positive correlation between these two features. This indicates that there might be a slight tendency for older individuals to work slightly more hours per week, but the correlation is not strong as shown in the polt above .
